```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Ramen.csv')
```

```python
df.info
df['Google Reviews'] = pd.to_numeric(df['Google Reviews'], errors='coerce')
```

```python
sns.regplot(data=df, x='Yelp Reviews', y='Google Reviews', scatter=True)
plt.locator_params(axis='y', nbins=10)
plt.title('Google vs Yelp Reviews')
plt.xlabel('Yelp Reviews')
plt.ylabel('Google Reviews')
plt.show
```

```python
correlation = df["Yelp Reviews"].corr(df['Google Reviews'])
determination = correlation ** 2
print("the correlation between x and y is",correlation,"and the determination is", determination)
```

```python
question=750/correlation
print(question)
```
