```python
import pandas as pd
import random as rd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
```

```python
amount = int(input("enter number of spins: "))
spin = []
for i in range(amount):
    spin.append(str(rd.randint(1,10)))
```

```python
count = Counter(spin)
df = pd.DataFrame.from_dict(count, orient='index', columns=['count'])
df = df.reset_index().rename(columns={'index': 'item'})
df = df.sort_values('count', ascending=False)
plt.bar(df['item'], df['count'], color='skyblue')
plt.title('Frequency Distribution')
plt.xlabel('number')
plt.ylabel('Frequency')
plt.tight_layout()
```
