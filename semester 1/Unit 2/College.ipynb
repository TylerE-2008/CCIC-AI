#  should i go to CCD and transfer to university of denver or go to uni of denver right after graduating if i want to maximize my expected net earnings in the first 5 years of a career in data science

# i am looking for profit after 5 years
# i will simulate one path that completes 4 year degree then gets a job
# how much money i profit from a degree

```python
CCannual = 9431 #community college debt
Cannual = 36037 #uni of denver debt
Ccash = 71155 #average money from denver
chance = 75 #graduation rate for denver
Nannual = 0 #no college debt
Ncash = 9504 #mcdonald cashier annual
CCcash = 56031 #associates pays
```

```python
import random as rd
import seaborn as sns
```

```python
Tprofit = 0
for i in range(100):
    if rd.randint(1,100) > 75:
        debt = Cannual * rd.randint(0,3)
        profit = debt - (Ncash * 5) * -1
    else:
        debt = Cannual * 4
        profit = debt - (rd.randint(60000, 80000) * 5) * -1
    Tprofit += profit
Tprofit = Tprofit / 100
print(Tprofit) #straight college
```

```python
Tprofit = 0
for i in range(100):
    if rd.randint(1,100) > 17:
        debt = CCannual * 1
        profit = debt - (Ncash * 5) * -1
    else:
        if rd.randint(1,100) < 75:
            debt = (CCannual * 2) + (Cannual * 2)
            profit = debt - (rd.randint(60000, 80000) * 5) * -1
        else:
            debt = (CCannual * 2) + Cannual
            profit = debt - (CCcash * 5) * -1
    Tprofit += profit
Tprofit = Tprofit / 100
print(Tprofit) #community college
```

# if you want to maximize profits after 5 years you want to go straight to college with 420k profit after debt than community college with a small chance to graduate and a 138k profit after debt
