# does the price of a book determine how much its rated, and how high
# the study for this is a prospective observational study with a full list of best sellers

# only the first 50 books were included for each year spanning 2009 - 2019.  Any data that couldnâ€™t be found in these lists was gathered using the GoodReads API
# a bias in this dataset is that there are repeating books from different years

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('books.csv')
df.info
```

```python
sns.regplot(data=df, x='Price', y='Reviews', scatter=True)
plt.xlim(0,30)
plt.locator_params(axis='y', nbins=10)
plt.title('Reviews vs Price')
plt.xlabel('Price')
plt.ylabel('Reviews')
plt.show
```

# looks like a lump around 7 to 15 price

```python
plt.hexbin(df['Reviews'], df['Price'], gridsize=30, cmap='Blues')
plt.colorbar(label='Count')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.title('Reviews vs Price')
plt.ylim(0, 30)
plt.show()
```

```python
correlation = df['Price'].corr(df['Reviews'])
print(correlation)
```

```python
sns.regplot(data=df, x='Price', y='User Rating', scatter=True)
plt.xlim(0,30)
plt.locator_params(axis='y', nbins=10)
plt.title('Reviews vs Price')
plt.xlabel('Price')
plt.ylabel('Rating')
plt.show
```

# from the correlation, hex graph, and scatter plot i believe that the price very slightly effects how many reviews it gets with the most being in a range of 5 to 17 dollars, but i do believe that price of the book has an impact on the ratings with books in the range 5 to 17 having a lower rating then other books.
