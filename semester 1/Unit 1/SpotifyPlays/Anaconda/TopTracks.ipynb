```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print(f"Pandas version: {pd.__version__}")
```

```python
df = pd.read_csv('songs.csv', encoding='ISO-8859-1')
```

```python
df.info
```

```python
counts = df["Artist"].value_counts()
counts = counts[counts > 15]
print(counts)
```

```python
sns.barplot(x=counts.index, y=counts.values)

plt.xticks(rotation = 90)
plt.title("Artist Frequency")
plt.xlabel("Artists")
plt.ylabel("Count")
plt.show() 
```

```python
relative = ((counts/counts.sum())*100)
```

```python
plt.figure(figsize = (8,8))
plt.pie(relative, labels=relative.index, autopct='%1.1f%%', startangle=140, pctdistance=0.85)
plt.title("Top Artists")
plt.axis("equal")
plt.tight_layout()
plt.show()
```

# the most frequent artists were Drake, Taylor Swift, and Bad Bunny, with 9.4%, 9.4%, and 9.0% respectively
